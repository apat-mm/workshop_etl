```python
import psycopg2
from pandas import DataFrame
import json
```

```python
with open('./db_config.json', 'r') as archivo:
    datos = json.load(archivo)
```

```python
try:
    conn = psycopg2.connect(
        host = 'localhost',
        user = datos['user'],
        password = datos['password'],
        database = datos['database']
    )
    print('¡Conexion exitosa!!')
    cursor = conn.cursor()
    
except Exception as ex:
    print(ex)
```

```python
sql = '''
    CREATE TABLE candidates
    (id SERIAL PRIMARY KEY,
    first_name varchar, 
    last_name varchar,
    email varchar,
    application_date date,
    country varchar,
    YOE varchar,
    seniority varchar,
    technology varchar,
    code_challenge_score INT,
    technical_interview_score INT
    );
       '''   
cursor.execute(sql)
conn.commit()
print('¡Tabla creada!')
```

```python
sql = '''
        COPY candidates(first_name, last_name, email, application_date, country, YOE, seniority, technology, code_challenge_score, technical_interview_score)
        FROM 'C:/datac/candidates.csv' DELIMITER ';' CSV HEADER;
          '''
cursor.execute(sql)
conn.commit()
print('¡Registros ingresados!')
```

```python
print('Primeros 3 registros de la tabla de candidates para evidenciar carga satisfactoria a la base de datos \n')
sql = '''
        SELECT * FROM candidates LIMIT 3
          '''
cursor.execute(sql)
conn.commit()

rows = cursor.fetchall()
columns = [col[0] for col in  cursor.description]
data = DataFrame(rows)
data = data.set_axis(columns, axis=1, inplace=False)
data
```

```python
sql = '''
            ALTER TABLE candidates
            ADD COLUMN setTech VARCHAR;
          '''
cursor.execute(sql)
conn.commit()
print('¡Columna setTech creada')
```

```python
print('Valores unicos de la variables technology \n')
sql = '''
        SELECT technology, COUNT(*) AS count
        FROM candidates GROUP BY technology;
          '''
cursor.execute(sql)
conn.commit()

rows = cursor.fetchall()
columns = [col[0] for col in  cursor.description]
data = DataFrame(rows)
data = data.set_axis(columns, axis=1, inplace=False)
data
```

```python
sql = '''
            UPDATE candidates 
            SET setTech = CASE
                WHEN technology IN ('Development - CMS Backend', 'Development - FullStack', 'Development - Backend', 'Development - Frontend', 'Development - CMS Frontend','Game Development') THEN 'Development'
                WHEN technology = 'Data Engineer' THEN 'Data'
                WHEN technology IN ('Salesforce', 'Ventas', 'Client Success') THEN 'Sales'
                WHEN technology IN ('QA Automation', 'QA Manual') THEN 'QA'
                WHEN technology IN ('DevOps', 'System Administration', 'Database Administration', 'Security Compliance', 'Security') THEN 'Security'
                WHEN technology IN ('Mulesoft', 'Adobe Experience Manager') THEN 'Technologies'
                WHEN technology IN ('Business Analytics / Project Management', 'Social Media Community Management') THEN 'Communication'
                ELSE 'Others'
            END;
          '''
cursor.execute(sql)
conn.commit()
print('Clasificacion de tecnologias')
```

```python
print('Consulta de contratados aprupados por sus especialidades: \n')
sql = '''
        SELECT setTech, COUNT(*) AS count
        FROM candidates WHERE hired IS TRUE GROUP BY setTech;
          '''
cursor.execute(sql)
conn.commit()

rows = cursor.fetchall()
columns = [col[0] for col in  cursor.description]
data = DataFrame(rows)
data = data.set_axis(columns, axis=1, inplace=False)
data
```

```python
print('Consulta de contratados agrupados por su nivel laboral: \n')
sql = '''
        SELECT seniority, COUNT(*) AS count
        FROM candidates WHERE hired IS TRUE GROUP BY seniority;
          '''
cursor.execute(sql)
conn.commit()

rows = cursor.fetchall()
columns = [col[0] for col in  cursor.description]
data = DataFrame(rows)
data = data.set_axis(columns, axis=1, inplace=False)
data
```

```python
print('Consulta de contratados agrupados por año: \n')
sql = '''
        SELECT EXTRACT(year FROM  application_date) as year, count(*) 
        FROM candidates WHERE hired IS TRUE GROUP BY year;

          '''
cursor.execute(sql)
conn.commit()

rows = cursor.fetchall()
columns = [col[0] for col in  cursor.description]
data = DataFrame(rows)
data = data.set_axis(columns, axis=1, inplace=False)
data
```

```python
print('Consulta de contratados en Colombia, Brasil, Ecuador y Estados Unnidos agrupado por año: \n')
sql = '''
        SELECT EXTRACT(year FROM  application_date) as year, count(*), country 
        FROM candidates WHERE hired IS TRUE AND (country = 'Colombia' OR country = 'Brazil' OR country='Ecuador' OR country='United States of America') GROUP BY year, country;

          '''
cursor.execute(sql)
conn.commit()

rows = cursor.fetchall()
columns = [col[0] for col in  cursor.description]
data = DataFrame(rows)
data = data.set_axis(columns, axis=1, inplace=False)
data
```
